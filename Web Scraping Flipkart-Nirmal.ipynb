```python
products=[]
prices=[]
specification=[]
ratings=[]
discount=[]
delivery=[]
Rating=[]
Ram=[]
OS=[]
storage=[]
Display=[]
OS_Type=[]
Warranty=[]
```

```python
for data in soup.findAll('div',class_='_3pLy-c row'):
    names=data.find('div', attrs={'class':'_4rR01T'})
    price=data.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
    ratings=data.find('div', attrs={'class':'gUuXy-'})
    specification = data.find('div', attrs={'class':'fMghEO'})
    discount=data.find('div',attrs={'class':'_3Ay6Sb'})
    delivery=data.find('div',attrs={'class':'_2Tpdn3'})
    print(names.text,price.text)
```

```python

for each in specification:
    col=each.find_all('li', attrs={'class':'rgWa7D'})
    
    ram=col[1].text
    os=col[2].text
    hard_disk=col[3].text
    display_size=col[4].text
    os_type=col[5].text
    warranty=col[6].text
    
products.append(names.text) 
prices.append(price.text) 
delivery.append(delivery.text)
discount.append(discount.text)
Rating.append(ratings)
Ram.append(ram)
OS.append(os)
storage.append(hard_disk)
Display.append(display_size)
OS_Type.append(os_type)
Warranty.append(warranty)
    
    
```

```python
print(len(products))
print(len(Rating))
print(len(prices))


```
